# Define the class of Binary Classification (digit)

```python
class BC:
    def __init__(self, trainmatrixX, trainmatrixY, testmatrixX, testmatrixY, featureCount, clustering):
        self.trainmatrixX = trainmatrixX
        self.trainmatrixY = trainmatrixY
        self.testmatrixX = testmatrixX
        self.testmatrixY = testmatrixY
        self.featureCount = featureCount
        self.clustering = clustering

    def train(self):
        FN = []; FP = []; TN = []; TP = []
        trainmatrixy = self.makevecY(self.trainmatrixY)
        testmatrixy = self.makevecY(self.testmatrixY)
        r = self.makeR()
        trainmatrixx = self.makevecX(self.trainmatrixX.T)
        testmatrixx = self.makevecX(self.testmatrixX.T)
        feature = self.makeFeature(r, trainmatrixx, self.featureCount)
        feature = np.mat(feature)
        weight = self.computeApproximation(feature, trainmatrixy)
        matrixResult = testmatrixx*weight
        FN, FP, TN, TP = self.binaryClassifier(testmatrixx, matrixResult, testmatrixy)
        self.computeAccuracy(FN, FP, TN, TP)
        self.visualization(FN, FP, TN, TP)
        
    def makevecY(self, matrix):
        for i in range(0, len(matrix)):
            if(int(matrix[i]) == int(self.clustering)):
                matrix[i] = 1 # label == 0, return 1
            else:
                matrix[i] = -1 # label != 0, return -1
        return matrix
    
    def makeR(self): 
        mu, sigma = 0, 0.1 # mean and standard deviation
        r = np.random.normal(mu, sigma, 784)
        return r

    def filt(self, matrix):
        kernel = np.array([[ 1, 4, 7, 4, 1],[4,  16,  26,  16, 4],[7, 26, 41, 26, 7], [4, 16, 26, 16, 4],[1, 4, 7, 4, 1]])
        matrix = signal.convolve2d(matrix, kernel, boundary='symm', mode='same')
        return matrix
    
    def makeFeature(self, feat, matrix, count):
        f_y = []
        for i in range(0, count):
            y = feat*matrix[i]
            f_y.append(y)
        return f_y

    def makevecX(self, makematrix):
        matrix = []
        for i in range(len(makematrix)):
            matrix_ = makematrix[i].reshape((size_row, size_col))
            matrix_ = self.filt(matrix_)
            matrix_ = np.ravel(matrix_, order='C')
            matrix.append(matrix_)
        return matrix
    
    def computeApproximation(self, vecX, vecY):
        vecX = np.mat(vecX)
        vecY = np.mat(vecY).T
        xTx = vecX.T * vecX
        if(np.linalg.det(xTx) == 0.0): # if Singular function, return 
            print("This matrix is singular")
            return
        weight = xTx.I * (vecX.T * vecY) # weight
        return weight
    
    def binaryClassifier(self, matrix, result, real):
        FN = []; FP = []; TN = []; TP = []
        for i in range(0, len(result)):
            if(int(result[i]) > 0):
                if(int(real[i]) == 1):
                    TP.append(matrix[i])
                else:
                    FP.append(matrix[i])
            else:
                if(int(real[i]) == 1):
                    FN.append(matrix[i])
                else:
                    TN.append(matrix[i])
        return FN, FP, TN, TP
    
    def computeAccuracy(self, FN, FP, TN, TP):
        accuracy = ((len(TN)+len(TP))/10000)*100
        print("Accuracy: ",accuracy)
        if((len(TP)+len(FP)) != 0):
            precision = 2*(len(TP))/(len(TP)+len(FP))
        else:
            precision = 0
        if((len(FN)+len(TP)) != 0):
            recall = len(TP)/(len(FN)+len(TP))
        else:
            recall = 0        
        if(float(precision+recall) != 0.0):
            f1score = 2*(precision*recall)/(precision+recall)
        else:
            f1score = 0
        print("\nf1score: ", f1score)
        
    def computeMin(self, vec):
        vec = np.mat(vec)
        mean_ = np.mean(vec, axis=0)
        return mean_
    
    def visualization(self, FN, FP, TN, TP):
        f2 = plt.figure(2)
        if(len(FN) != 0):
            im_average  = np.zeros((size_row * size_col, 10), dtype=float)
            im_count    = np.zeros(10, dtype=int)
            P1 = plt.subplot(2, 2, 1)
            P1.set_title('False Negative')
            im_average = self.computeMin(FN)
            plt.imshow(im_average.reshape((size_row, size_col)), cmap='Greys', interpolation='None') 
            plt.axis('off')
        
        if(len(TN) != 0):
            P2 = plt.subplot(2, 2, 2)
            P2.set_title('True Negative')
            im_average = self.computeMin(TN)
            plt.imshow(im_average.reshape((size_row, size_col)), cmap='Greys', interpolation='None') 
            plt.axis('off')

        if(len(FP) != 0):
            P3 = plt.subplot(2, 2, 3)
            P3.set_title('False Positive')
            im_average = self.computeMin(FP)
            plt.imshow(im_average.reshape((size_row, size_col)), cmap='Greys', interpolation='None') 
            plt.axis('off')
        
        if(len(TP) != 0):
            P4 = plt.subplot(2, 2, 4)
            P4.set_title('True Positive')
            im_average = self.computeMin(TP)
            plt.imshow(im_average.reshape((size_row, size_col)), cmap='Greys', interpolation='None') 
            plt.axis('off')
```
